```python
#¿Cuántos desocupados hay en la muestra? ¿Cuántos inactivos?

#variable estado: condición de actividad
#0 = Entrevista individual no realizada (no respuesta al cuestionario individual)
#1 = Ocupado
#2 = Desocupado
#3 = Inactivo
#4 = Menor de 10 años

#contamos los desocupados (valor 2 en la variable estado)
desocupados = df_limpio2['estado'].value_counts().get(2, 0)  #2 representa desocupados
print(f"Cantidad de desocupados en la muestra: {desocupados}")

#contamos los inactivos (valor 3 en la variable estado)
inactivos = df_limpio2['estado'].value_counts().get(3, 0)  #3 representa inactivos
print(f"Cantidad de inactivos en la muestra: {inactivos}")
```

```python
#vamos a agregar una columna llamada PEA a la base respondieron que será 1 si el estado es 1 u 2 (ocupado, desocupado) y 0 en caso contrario

import matplotlib.pyplot as plt

#agregamosla columna PEA
#1 si está ocupado (estado = 1) o desocupado (estado = 2), 0 en otros casos
respondieron['PEA'] = respondieron['estado'].apply(lambda x: 1 if x in [1, 2] else 0)
#print(respondieron['PEA']) chequeamos que este bien

#filtramos los datos por años 2004 y 2024
respondieron_2004 = respondieron[respondieron['ano4'] == 2004]
respondieron_2024 = respondieron[respondieron['ano4'] == 2024]

#contamos la cantidad de personas en PEA para 2004 y 2024
pea_2004 = respondieron_2004['PEA'].value_counts()
pea_2024 = respondieron_2024['PEA'].value_counts()
#print(pea_2004) chequeamos que este bien
#print(pea_2024) chequeamos que este bien

#creamos el gráfico de barras comparando los dos años
#0: PEI (Población Económicamente Inactiva), 1: PEA (Población Económicamente Activa)
labels = ['PEI 2004', 'PEA 2004', 'PEI 2024', 'PEA 2024']
valores = [pea_2004[0], pea_2004[1], pea_2024[0], pea_2024[1]]
fig, ax = plt.subplots()

ax.bar(labels, valores, color=['blue', 'green', 'blue', 'green'])

ax.set_ylabel('Número de personas')
ax.set_title('Composición de la Población Económicamente Activa (PEA) y PEI en 2004 y 2024')
plt.show()

#COMENTARIOS
# - en ambos años el PEI (azul) es mayor al PEA (verde)
# - en 2004 la diferencia entre PEA y PEI es menor que en 2024
# - en 2024 hay menos cantidad de sujetos 
```

```python
#vamos a  agregar una columna llamada PET (Población en Edad para Trabajar) que tome 1 si están la persona tiene entre 15 y 65 años cumplidos

#personas entre 15 y 65 años: aquellas cuyo valor de ch06 (edad) esta entre 15 y 65
#personas mas chicas entre 15 y mayores a 65 años: aquellas cuyo valor de ch06 (edad) esta por fuera de 15 y 65

#agregamosla columna PET
#1 si la persona tiene entre 15 y 65 años cumplidos, 0 en otros casos
respondieron['PET'] = respondieron['ch06'].apply(lambda x: 1 if 15 <= x <= 65 else 0)

#filtramos los datos por años 2004 y 2024
respondieron_2004 = respondieron[respondieron['ano4'] == 2004]
respondieron_2024 = respondieron[respondieron['ano4'] == 2024]

#contamos la cantidad de personas en PET para 2004 y 2024
pet_2004 = respondieron_2004['PET'].value_counts()
pet_2024 = respondieron_2024['PET'].value_counts()
#print(pet_2004) chequeamos que este bien
#print(pet_2024) chequeamos que este bien

#creamos el gráfico de barras comparando los dos años
#0: SET (Población Sin Edad para Trabajar), 1: PET (Población en Edad para Trabajar)
labels = ['SET 2004', 'PET 2004', 'SET 2024', 'PET 2024']
valores = [pet_2004[0], pet_2004[1], pet_2024[0], pet_2024[1]]
fig, ax = plt.subplots()

ax.bar(labels, valores, color=['red', 'yellow', 'red', 'yellow'])

ax.set_ylabel('Número de personas')
ax.set_title('Composición de la Población en Edad para Trabajar (PET) y SET en 2004 y 2024')
plt.show()

#COMENTARIOS
# - en ambos años hay menos personas sin edad para trabajar (rojo)
# - las diferencias en SET y PET son similares en los dos años, es decir, parece haber una proporcion similar 
# - hay menos datos en 2024 
```

```python
#AÑO 2004

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, accuracy_score
import matplotlib.pyplot as plt

#planteamos el modelo de regresion logistica
reglog_2004 = LogisticRegression(solver='liblinear')

#entrenamos el modelo con los datos de entrenamiento 
reglog_2004.fit(X_train_2004, Y_train_2004)

#hacemos las predicciones con los datos de prueba
y_pred_2004_reglog = reglog_2004.predict(X_test_2004)
y_prob_2004_reglog = reglog_2004.predict_proba(X_test_2004)[:, 1]

#matriz de confusion
matriz_2004_reglog = confusion_matrix(Y_test_2004, y_pred_2004_reglog)
print("Matriz de Confusión 2004:\n", matriz_2004_reglog)

#accuracy
accuracy_2004_reglog = accuracy_score(Y_test_2004, y_pred_2004_reglog)
print(f"Accuracy 2004: {accuracy_2004_reglog}")

#AUC
auc_2004_reglog = roc_auc_score(Y_test_2004, y_prob_2004_reglog)
print(f"AUC 2004: {auc_2004_reglog}")

#curva ROC
fpr_2004_reglog, tpr_2004_reglog, _ = roc_curve(Y_test_2004, y_prob_2004_reglog)
plt.figure()
plt.plot(fpr_2004_reglog, tpr_2004_reglog, label=f"ROC Curve 2004 (AUC = {auc_2004_reglog:.2f})")
plt.plot([0, 1], [0, 1], 'k--')  #linea diagonal
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Curva ROC 2004')
plt.legend(loc='best')
plt.show()
```
